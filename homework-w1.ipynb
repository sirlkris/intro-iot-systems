```python
#Kristen Sirel
#
import pandas as pd 
import matplotlib.pyplot as plt 


Temperature = pd.read_csv("FMI_lpr_2019.csv",dayfirst=True,sep=",", 
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])
Temperature.info()

#Year plot
plt.figure(figsize=(30,15)) 
plt.plot(Temperature,color='blue',linestyle='-') 
plt.title("Air temperature in 2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.grid(True) 
plt.show() 

#Month plot
plt.figure(figsize=(16,6)) 
plt.plot(Temperature['2019-1'],color='blue',linestyle='-') 
plt.title("Air temperature in January in 2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.grid(True) 
plt.show() 

#Week plot
plt.figure(figsize=(16,6)) 
plt.plot(Temperature['2019-1-7':'2019-1-13'],color='blue',linestyle='-') 
plt.title("Air temperature in week 2 of 2019,7-13.1.2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.grid(True) 
plt.show() 


#Day plot
plt.figure(figsize=(16,6)) 
plt.plot(Temperature['2019-1-14'],color='blue', marker='o',linestyle='-') 
plt.title("Air temperature in 14th of January, 2018, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.grid(True) 
plt.show() 

#eof
```
