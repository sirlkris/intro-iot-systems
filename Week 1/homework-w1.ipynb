```python
#Kristen Sirel
#
import pandas as pd 
import matplotlib.pyplot as plt 


Temp = pd.read_csv("FMI_lpr_2019.csv",dayfirst=True,sep=",", 
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])
Temp.info()

#Year plot
plt.figure(figsize=(30,15)) 
plt.plot(Temp,color='blue',linestyle='-') 
plt.title("Air temperature in 2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.xlabel("Time in [h]")
plt.grid(True) 
plt.show() 

#Month plot
plt.figure(figsize=(16,6)) 
plt.plot(Temp['2019-1'],color='blue',linestyle='-') 
plt.title("Air temperature in January in 2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.xlabel("Time in [h]")
plt.grid(True) 
plt.show() 

#Week plot
plt.figure(figsize=(16,6)) 
plt.plot(Temp['2019-1-7':'2019-1-13'],color='blue',linestyle='-') 
plt.title("Air temperature in week 2 of 2019,7-13.1.2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.xlabel("Time in [h]")
plt.grid(True) 
plt.show() 


#Day plot
plt.figure(figsize=(16,6)) 
plt.plot(Temp['2019-1-14'],color='blue', marker='o',linestyle='-') 
plt.title("Air temperature in 14th of January, 2019, at Lappeenranta airport station") 
plt.ylabel("Temperature in [째C]") 
plt.xlabel("Time in [h]") 
plt.grid(True) 
plt.show() 

#eof
```

**(4) Consider a house with an electric heater. How the electricity consumption of the heater would be affected by the air temperature? Is there a way to change this?**

Answer:
If the heater is temperature controled, colder air temerature means that the heater has to be on more often or longer times compared to warmer air temperature. This makes the heater to consume more eletricity. If the air is warmer/close to the desired temerature, time that the heater has to be on reduces. At the same time eletricity comsumption is reduced. By chaning control type of the heater, the heater can be changet to output constant heat, so it doesnt follow air tempertautre thus consuming constant eletrical energy.
